```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```

df=pd.read_csv('/content/2023_Countries by Population.csv')
```

```
df.head()
```

```
df.tail()
```

```
# Set the index of the DataFrame to the 'country' column
population_data = df.set_index('country')
```

```
population_data.columns
```

```
population_data.dtypes
```

```
# Convert 'place' column to categorical data type
population_data['place'] = population_data['place'].astype('category')

# Convert 'cca2' column to categorical data type
population_data['cca2'] = population_data['cca2'].astype('category')

# Convert 'cca3' column to categorical data type
population_data['cca3'] = population_data['cca3'].astype('category')
```

```
population_data.describe()
```

```
import matplotlib.pyplot as plt

# Select the top 10 countries by population in 2022
top10 = population_data.sort_values(by='pop2022', ascending=False).head(10)

# Create a bar chart of the top 10 countries by population in 2022
plt.bar(top10.index, top10['pop2022'])
plt.xticks(rotation=90)
plt.ylabel('Population in 2022')
plt.title('Top 10 Countries by Population in 2022')
plt.show()
```

```
# Create a scatter plot of population density vs. land area for all countries
plt.scatter(population_data['landAreaKm'], population_data['density'])
plt.xscale('log')
plt.yscale('log')
plt.xlabel('Land Area (sq. km)')
plt.ylabel('Population Density (people/sq. km)')
plt.title('Population Density vs. Land Area for All Countries')
plt.show()
```

```
# Select a country to plot
country = 'China'

# Create a line chart of population growth rates over time for the selected country
plt.plot(population_data.loc[country, 'pop1980':'pop2050'], '-o')
plt.xlabel('Year')
plt.ylabel('Population Growth Rate (%)')
plt.title(f'Population Growth Rates Over Time for {country}')
plt.show()
```
